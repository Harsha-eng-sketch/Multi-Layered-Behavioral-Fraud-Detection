```python
def validate_data(df: pd.DataFrame) -> bool:
    """
    Validate that data meets quality checks
    
    Args:
        df: Input DataFrame
        
    Returns:
        True if validation passes
    """
    checks_passed = True
    
    # Check for missing values
    missing_cols = df.columns[df.isnull().any()].tolist()
    if missing_cols:
        print(f"⚠️ Missing values found in: {missing_cols}")
        checks_passed = False
    
    # Check target column exists
    if 'target_fraud' not in df.columns:
        print("⚠️ 'target_fraud' column not found!")
        checks_passed = False
    
    # Check required columns
    required_cols = [ 
        'customer_id', 'age', 'gender', 'country', 'tenure_months',
        'monthly_spend', 'num_transactions', 'last_login_days',
        'support_tickets', 'payment_method', 'is_active',
        'total_spend', 'avg_transaction_value'
    ]
     
     
    missing_required = [col for col in required_cols if col not in df.columns]
    if missing_required:
        print(f"⚠️ Missing required columns: {missing_required}")
        checks_passed = False
    
    if checks_passed:
        print("✅ Data validation passed!")
    
    return checks_passed

print("✅ validate_data() function defined")
```

```python
def add_spend_per_transaction(df: pd.DataFrame) -> pd.DataFrame:
    """
    Calculate average spend per transaction
    
    Formula: total_spend / (num_transactions + epsilon)
    Epsilon prevents division by zero
    """
    epsilon = 1e-5
    df['spend_per_transaction'] = df['total_spend'] / (df['num_transactions'] + epsilon)
    return df

print("✅ add_spend_per_transaction() defined")
```

```python
def add_login_frequency(df: pd.DataFrame) -> pd.DataFrame:
    """
    Calculate login frequency (inverse of days since last login)
    
    Formula: 1 / (last_login_days + 1)
    Higher value = more frequent logins
    """
    df['login_frequency'] = 1 / (df['last_login_days'] + 1)
    return df

print("✅ add_login_frequency() defined")
```

```python
def add_support_rate(df: pd.DataFrame) -> pd.DataFrame:
    """
    Calculate support tickets per month of tenure
    
    Formula: support_tickets / (tenure_months + 1)
    """
    df['support_rate'] = df['support_tickets'] / (df['tenure_months'] + 1)
    return df

print("✅ add_support_rate() defined")
```

```python
def add_burst_activity(df: pd.DataFrame) -> pd.DataFrame:
    """
    Calculate burst activity ratio (bonus feature)
    
    Formula: num_transactions / (last_login_days + 1)
    Detects suspicious burst of activity after dormant period
    """
    df['burst_activity'] = df['num_transactions'] / (df['last_login_days'] + 1)
    return df

print("✅ add_burst_activity() defined")
```

```python
def compute_peer_group_stats(df: pd.DataFrame, 
                             group_cols: List[str] = ['country', 'payment_method'],
                             feature_cols: List[str] = ['monthly_spend', 'num_transactions', 'avg_transaction_value']) -> Dict:
    """
    Compute mean and std for peer groups
    
    Args:
        df: Input DataFrame
        group_cols: Columns to group by (country, payment_method)
        feature_cols: Features to compute statistics for
        
    Returns:
        Dictionary containing group statistics
    """
    group_stats = {}
    
    for feature in feature_cols:
        # Calculate mean and std per group
        stats = df.groupby(group_cols)[feature].agg(['mean', 'std']).to_dict('index')
        group_stats[feature] = stats
    
    return group_stats

print("✅ compute_peer_group_stats() defined")
```

```python
def add_peer_deviation(df: pd.DataFrame, 
                       group_stats: Optional[Dict] = None,
                       group_cols: List[str] = ['country', 'payment_method'],
                       feature_cols: List[str] = ['monthly_spend', 'num_transactions', 'avg_transaction_value']) -> Tuple[pd.DataFrame, Dict]:
    """
    Add z-score features showing deviation from peer group
    
    Args:
        df: Input DataFrame
        group_stats: Pre-computed statistics (if None, computes from df)
        group_cols: Columns to group by
        feature_cols: Features to compute deviation for
        
    Returns:
        DataFrame with new z-score columns, and group_stats dictionary
    """
    if group_stats is None:
        group_stats = compute_peer_group_stats(df, group_cols, feature_cols)
    
    # Create mapping dictionaries for quick lookup
    for feature in feature_cols:
        z_score_col = f'z_{feature}'
        df[z_score_col] = 0.0
        
        # Apply z-score calculation per row
        for idx, row in df.iterrows():
            # Create group key tuple
            group_key = tuple(row[col] for col in group_cols)
            
            # Get stats for this group
            if group_key in group_stats[feature]:
                mean = group_stats[feature][group_key]['mean']
                std = group_stats[feature][group_key]['std']
                
                # Avoid division by zero
                if std > 0:
                    df.loc[idx, z_score_col] = (row[feature] - mean) / std
                else:
                    df.loc[idx, z_score_col] = 0
            else:
                # If group not found, set to 0 (no deviation)
                df.loc[idx, z_score_col] = 0
    
    return df, group_stats

print("✅ add_peer_deviation() defined")
```

```python
class FraudFeatureEngineer:
    """
    Reusable feature engineering pipeline for fraud detection
    
    Methods:
        fit(): Learn statistics from training data
        transform(): Apply feature engineering to any dataset
        fit_transform(): Fit and transform in one step
    """
    
    def __init__(self):
        self.scaler = StandardScaler()
        self.group_stats = None
        self.feature_columns = None
        self.categorical_cols = ['gender', 'country', 'payment_method']
        self.numerical_cols = [
            'age', 'tenure_months', 'monthly_spend', 'num_transactions',
            'last_login_days', 'support_tickets', 'total_spend',
            'avg_transaction_value', 'suspicion_score'
        ]
        self.engineered_cols = [
            'spend_per_transaction', 'login_frequency', 'support_rate',
            'burst_activity', 'z_monthly_spend', 'z_num_transactions',
            'z_avg_transaction_value'
        ]
        self.is_fitted = False
    
    def _add_all_features(self, df: pd.DataFrame) -> pd.DataFrame:
        """Apply all feature engineering transformations"""
        df = df.copy()
        
        # Basic features
        df = add_spend_per_transaction(df)
        df = add_login_frequency(df)
        df = add_support_rate(df)
        df = add_burst_activity(df)
        
        # Peer deviation features
        df, self.group_stats = add_peer_deviation(df, self.group_stats)
        
        return df
    
    def _preprocess_categorical(self, df: pd.DataFrame) -> pd.DataFrame:
        """One-hot encode categorical variables"""
        df = df.copy()
        
        # One-hot encode categorical columns
        for col in self.categorical_cols:
            dummies = pd.get_dummies(df[col], prefix=col, drop_first=False)
            df = pd.concat([df, dummies], axis=1)
            df = df.drop(columns=[col])
        
        return df
    
    def _get_feature_columns(self, df: pd.DataFrame, exclude_cols: List[str]) -> List[str]:
        """Get list of feature columns (exclude target and ID columns)"""
        return [col for col in df.columns if col not in exclude_cols]

print("✅ FraudFeatureEngineer class initialized")
```

```python
# Add fit method to FraudFeatureEngineer
def fit(self, df: pd.DataFrame, target_col: str = 'target_fraud'):
    """
    Fit the feature engineering pipeline
    
    Args:
        df: Training DataFrame
        target_col: Name of target column
    """
    # Add all features (will also compute group_stats)
    df_featured = self._add_all_features(df)
    
    # One-hot encode categoricals
    df_processed = self._preprocess_categorical(df_featured)
    
    # Define columns to exclude from features
    exclude_cols = [target_col, 'customer_id']
    
    # Get feature columns
    self.feature_columns = self._get_feature_columns(df_processed, exclude_cols)
    
    # Fit scaler on training data
    X_train = df_processed[self.feature_columns].values
    self.scaler.fit(X_train)
    
    self.is_fitted = True
    print(f"✅ FraudFeatureEngineer fitted successfully!")
    print(f"📊 Number of features: {len(self.feature_columns)}")
    print(f"🔧 Features: {self.feature_columns[:5]}... (showing first 5)")
    
    return self

# Attach method to class
FraudFeatureEngineer.fit = fit
print("✅ fit() method added to FraudFeatureEngineer")
```

```python
def transform(self, df: pd.DataFrame, target_col: str = 'target_fraud') -> Tuple[np.ndarray, Optional[np.ndarray]]:
    """
    Transform data using fitted pipeline
    
    Args:
        df: DataFrame to transform
        target_col: Name of target column
        
    Returns:
        X: Feature matrix (numpy array)
        y: Target vector (if available, else None)
    """
    if not self.is_fitted:
        raise RuntimeError("Must call fit() before transform()")
    
    # Add all features
    df_featured = self._add_all_features(df)
    
    # One-hot encode categoricals
    df_processed = self._preprocess_categorical(df_featured)
    
    # Ensure all feature columns exist (add missing with zeros)
    for col in self.feature_columns:
        if col not in df_processed.columns:
            df_processed[col] = 0
    
    # Extract features in correct order
    X = df_processed[self.feature_columns].values
    
    # Scale features
    X_scaled = self.scaler.transform(X)
    
    # Extract target if available
    y = None
    if target_col in df.columns:
        y = df[target_col].values
    
    return X_scaled, y

FraudFeatureEngineer.transform = transform
print("✅ transform() method added to FraudFeatureEngineer")
```

```python
def fit_transform(self, df: pd.DataFrame, target_col: str = 'target_fraud') -> Tuple[np.ndarray, np.ndarray]:
    """Fit and transform in one step"""
    self.fit(df, target_col)
    X, y = self.transform(df, target_col)
    return X, y

def save(self, artifact_dir: str = 'artifacts'):
    """Save fitted pipeline artifacts"""
    if not self.is_fitted:
        raise RuntimeError("Cannot save unfitted pipeline")
    
    artifact_path = Path(artifact_dir)
    artifact_path.mkdir(exist_ok=True)
    
    # Save artifacts
    joblib.dump(self.scaler, artifact_path / 'scaler.pkl')
    joblib.dump(self.group_stats, artifact_path / 'group_stats.pkl')
    joblib.dump(self.feature_columns, artifact_path / 'feature_columns.pkl')
    
    print(f"✅ Artifacts saved to {artifact_dir}/")
    print(f"   - scaler.pkl")
    print(f"   - group_stats.pkl")
    print(f"   - feature_columns.pkl")

def load(self, artifact_dir: str = 'artifacts'):
    """Load fitted pipeline artifacts"""
    artifact_path = Path(artifact_dir)
    
    self.scaler = joblib.load(artifact_path / 'scaler.pkl')
    self.group_stats = joblib.load(artifact_path / 'group_stats.pkl')
    self.feature_columns = joblib.load(artifact_path / 'feature_columns.pkl')
    self.is_fitted = True
    
    print(f"✅ Artifacts loaded from {artifact_dir}/")
    return self

# Attach methods to class
FraudFeatureEngineer.fit_transform = fit_transform
FraudFeatureEngineer.save = save
FraudFeatureEngineer.load = load

print("✅ fit_transform(), save(), and load() methods added to FraudFeatureEngineer")
```

```python
def prepare_training_data(df: pd.DataFrame, 
                         test_size: float = 0.2,
                         random_state: int = 42) -> Dict:
    """
    Complete pipeline: feature engineering + train/validation split
    
    Args:
        df: Raw DataFrame
        test_size: Validation set proportion
        random_state: Random seed for reproducibility
        
    Returns:
        Dictionary containing X_train, X_val, y_train, y_val, and feature_engineer
    """
    print("=" * 50)
    print("🔧 FEATURE ENGINEERING PIPELINE")
    print("=" * 50)
    
    # Initialize feature engineer
    fe = FraudFeatureEngineer()
    
    # Fit and transform
    X, y = fe.fit_transform(df, target_col='target_fraud')
    
    # Split into train/validation (stratified)
    X_train, X_val, y_train, y_val = train_test_split(
        X, y,
        test_size=test_size,
        random_state=random_state,
        stratify=y  # Maintain fraud ratio
    )
    
    print(f"\n📊 Data Split Summary:")
    print(f"   Training set: {X_train.shape[0]} samples, {X_train.shape[1]} features")
    print(f"   Validation set: {X_val.shape[0]} samples, {X_val.shape[1]} features")
    print(f"   Training fraud rate: {y_train.mean():.2%}")
    print(f"   Validation fraud rate: {y_val.mean():.2%}")
    
    # Save artifacts
    fe.save()
    
    # Save numpy arrays
    artifact_dir = Path('artifacts')
    np.save(artifact_dir / 'X_train.npy', X_train)
    np.save(artifact_dir / 'X_val.npy', X_val)
    np.save(artifact_dir / 'y_train.npy', y_train)
    np.save(artifact_dir / 'y_val.npy', y_val)
    
    print(f"\n💾 Saved numpy arrays to {artifact_dir}/")
    print(f"   - X_train.npy, X_val.npy")
    print(f"   - y_train.npy, y_val.npy")
    
    return {
        'X_train': X_train,
        'X_val': X_val,
        'y_train': y_train,
        'y_val': y_val,
        'feature_engineer': fe
    }

print("✅ prepare_training_data() defined")
```
